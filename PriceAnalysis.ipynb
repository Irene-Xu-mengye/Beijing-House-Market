```python
import pandas as pd
import numpy as np;
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/data.csv", delimiter = ",")
df_smaller = pd.read_csv("data/data_smaller.csv", delimiter = ",")
```

```python
house_quality = df[['square', 'bedRoom', 'livingRoom', 'kitchen', 'bathRoom', 'floor', 'constructionTime', 'ladderRatio', 'elevator','price', 'totalPrice']]
```

```python
loc_quality = df[['fiveYearsProperty', 'buildingType', 'buildingStructure',  'communityAverage', 'subway', 'district', 'totalPrice', 'price']]
```

```python
cor_house = house_quality.corr()
sns.heatmap(cor_house, cmap = "Greens")
plt.title("Correlation Heatmap of House Quality")
```

```python
cor_loc = loc_quality.corr()
sns.heatmap(cor_loc, cmap = "Greens")
plt.title("Correlation Heatmap of Location Quality")
```
